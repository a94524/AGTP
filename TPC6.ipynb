### Plot1: uma espécie de "Hello world!"

```python
# importing the required module
import matplotlib.pyplot as plt
 
# x axis values
x = [1,2,3]
# corresponding y axis values
y = [2,4,1]
 
# plotting the points
plt.plot(x, y)
 
# naming the x axis
plt.xlabel('Abcissas')
# naming the y axis
plt.ylabel('Ordenadas')
 
# giving a title to my graph
plt.title('O meu primeiro gráfico!')
 
# function to show the plot
plt.show()
```

### Plot2: duas funções no mesmo gráfico

```python
import matplotlib.pyplot as plt
 
# line 1 points
x1 = [1,2,3]
y1 = [2,4,1]
# plotting the line 1 points
plt.plot(x1, y1, label = "linha 1")
 
# line 2 points
x2 = [1,2,3]
y2 = [4,1,3]
# plotting the line 2 points
plt.plot(x2, y2, label = "linha 2")
 
# naming the x axis
plt.xlabel('Abcissas')
# naming the y axis
plt.ylabel('Ordenadas')
# giving a title to my graph
plt.title('Gráfico com duas funções')
 
# show a legend on the plot
plt.legend()
 
# function to show the plot
plt.show()
```

### Plot3: configurações

```python
import matplotlib.pyplot as plt
 
# x axis values
x = [1,2,3,4,5,6]
# corresponding y axis values
y = [2,4,1,5,2,6]
 
# plotting the points
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)
 
# setting x and y axis range
plt.ylim(1,8)
plt.xlim(1,8)
 
# naming the x axis
plt.xlabel('Abcissas')
# naming the y axis
plt.ylabel('Ordenadas')
 
# giving a title to my graph
plt.title('Um gráfico com estilo!')
 
# function to show the plot
plt.show()
```

### Exercício 1: desenha um quadrado centrado na origem com 5 unidades de lado

```python
import matplotlib.pyplot as plt

x=[-2.5,2.5,2.5,-2.5,-2.5]
y=[2.5,2.5,-2.5,-2.5,2.5]
plt.plot(x,y,label='Quadrado')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Quadrado centrado na origem de lado 5')

eixoXX=[-5,5]
eixoXY=[0,0]
plt.plot(eixoXX,eixoXY,label='Eixo do X')

eixoXY=[0,0]
eixoYY=[-5,5]
plt.plot(eixoXY,eixoYY,label='Eixo do Y')
plt.show()
```

## Gráficos de barras

```python
import matplotlib.pyplot as plt
 
# x-coordinates of left sides of bars
left = [1, 2, 3, 4, 5]
 
# heights of bars
height = [6, 4, 8, 12, 5]
 
# labels for bars
tick_label = ['cebolas', 'maçãs', 'pêras', 'batatas', 'cenouras']
 
# plotting a bar chart
plt.bar(left, height, tick_label = tick_label,
        width = 0.8, color = ['red', 'green'])
 
# naming the x-axis
plt.xlabel('Frutas e legumes')
# naming the y-axis
plt.ylabel('Quantidade')
# plot title
plt.title('O meu stock de verdes')
 
# function to show the plot
plt.show()
```

## Gráficos em forma de tarte

```python
import matplotlib.pyplot as plt
 
# defining labels
activities = ['eat', 'sleep', 'work', 'play']
 
# portion covered by each label
slices = [3, 7, 8, 6]
 
# color for each label
colors = ['r', 'y', 'g', 'b']
 
# plotting the pie chart
plt.pie(slices, labels = activities, colors=colors,
        startangle=90, shadow = True, explode = (0, 0, 0.1, 0),
        radius = 1.2, autopct = '%1.1f%%')
 
# plotting legend
plt.legend()
 
# showing the plot
plt.show()
```
